# ETL

```python
import data, imp
imp.reload(data)
from data import load_method_dict, load_method_dict_multiprocessing

def _etl_dataset(inp):
    
    exp_dir, dataset_id, q_list, fold_ident, trte_idx, methods = inp
    print(dataset_id)
    
    method_to_etls = dict()
    dir_files = get_dir_files(exp_dir, dataset_id)
    method_dict = load_method_dict(dir_files, fold_ident)
    
    if methods is None:
        methods = method_dict[0]
    else:
        methods = [method for method in methods if method in method_dict[0]] 
        
    n_folds = len(method_dict)
    for method in methods:

        etls_ = []
        for fold in range(n_folds):
            test_normed_residuals = np.abs(method_dict[fold][method][trte_idx]['pred_residual_normed'].values) # shape: len(fold)
            q_val = np.quantile(test_normed_residuals, q_list) # shape: len(q_list)
            
            mask = test_normed_residuals[None, ...] <= q_val[..., None]
            filtered_residuals = np.stack([test_normed_residuals for i in range(len(q_list))])
            filtered_residuals[mask] = np.nan
            
            etls = np.nanmean(filtered_residuals, axis=1) # shape: len(q_list)
            etls_.append(etls)

        method_to_etls[method] = etls_
        
    return method_to_etls

def etl(exp_dirs, q_list, fold_ident='random_folds', trte_idx=1, methods=None, datasets=None):
    
    res = []
    for exp_dir in exp_dirs:
        for dataset_id in os.listdir(exp_dir):
            if dataset_id in available_datasets and (datasets is None or dataset_id in datasets) :
                 res.append(_etl_dataset((exp_dir, dataset_id, q_list, fold_ident, trte_idx, methods)))
    
    method_to_etls = dict()
    for dataset_res in res:
        for method in dataset_res:
            if method not in method_to_etls:
                method_to_etls[method] = []
            method_to_etls[method].append(dataset_res[method])
    
    aggregated_etls = dict()
    for method in method_to_etls:
        etls = np.concatenate(method_to_etls[method]) # method_to_etls[method] shape: n_datasets x n_folds x len(q_list)
        method_to_etls[method] = etls
        aggregated_etls[method] = np.array([np.nanmean(etls, axis=0), np.nanmax(etls, axis=0)])
    
    for i, q in enumerate(q_list):
        for method in method_to_etls.keys():
            print(method, aggregated_etls[method][0, i], aggregated_etls[method][1, i])
        
    return method_to_etls

exp_dirs = ['INSERT_PATHS_HERE'] # TODO: INSERT PATHS HERE

etl(exp_dirs, [0.99], fold_ident='random_folds', methods=['new_wdrop_exact_l=5', 'mc_pu', 'pu_de'])
etl(exp_dirs, [0.99], fold_ident='label_folds', methods=['new_wdrop_exact_l=5', 'mc_pu', 'pu_de'])
etl(exp_dirs, [0.99], fold_ident='pca_folds', methods=['new_wdrop_exact_l=5', 'mc_pu', 'pu_de'])
```
